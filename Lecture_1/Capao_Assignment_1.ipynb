```python
from ucimlrepo import fetch_ucirepo
import numpy as numpy
  
zoo = fetch_ucirepo(id=111) 
  
X = zoo.data.features 
y = zoo.data.targets    
 
print("Chosen Data set is ZOO\n\n") 
print(zoo.metadata) 
print(zoo.variables)
print("---")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
cmap = sns.diverging_palette(0, 255, sep=30, n=256, as_cmap=True)
sns.heatmap(X.corr(), annot=True, cmap=cmap)
plt.suptitle("Kalyr's", color='mediumvioletred', fontsize=20)
plt.title('Correlation Plot of ZOO features', fontsize=30, fontweight='bold')
plt.show()
```

$\textbf{Graphs of Animal Features}$

```python
import scipy.stats as stats

for column in X.columns:
    plt.figure(figsize=(4, 4))
    X[column].hist(bins=10, color='pink')
    plt.title(f'Zoo animals that has {column}')
    plt.xlabel(column)
    plt.ylabel('number of animals')
    plt.show()
```
