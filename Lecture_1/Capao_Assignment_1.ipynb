```python
from ucimlrepo import fetch_ucirepo
import numpy as numpy
  
zoo = fetch_ucirepo(id=111) 
  
X = zoo.data.features 
y = zoo.data.targets    
 
print("Chosen Data set is ZOO\n\n") 
print(zoo.metadata) 
print(zoo.variables)
print("---")
```

$\textbf{Chosen Dataset:}$
https://archive.ics.uci.edu/dataset/111/zoo 

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(X.corr(), dtype=bool))
cmap = sns.diverging_palette(0, 255, sep=30, n=256, as_cmap=True)
sns.heatmap(X.corr(), annot=True, cmap=cmap, mask=mask)
plt.suptitle("Kalyr's", color='mediumvioletred', fontsize=20)
plt.title('Correlation Plot of ZOO features', fontsize=30, fontweight='bold')
plt.show()
```

$\textbf{Summary Statistics}$

```python
from tabulate import tabulate
import scipy.stats as stats
import matplotlib.pyplot as plt
summary_stats = X.describe()
summary_stats = summary_stats.T
print(summary_stats)
```

$\textbf{Distribution of Animal Features}$

```python
for column in X.columns:
    plt.figure(figsize=(4, 4))
    sns.histplot(X[column], bins='auto', color='pink', kde=True, cbar=True)
    plt.title(f'Distribution of Zoo animals that has {column}')
    plt.xlabel(column)
    plt.ylabel('number of animals')
    plt.show()
```
